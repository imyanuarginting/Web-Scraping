```python
from googleapiclient.discovery import build

API_SERVICE_NAME = 'youtube'
API_VERSION = 'v3'
API_KEY = 'AIzaSyAaAks13A2icJ_rBRxEc_ESIZn2q9Wf8hQ'
VIDEO_ID = 'k0ubS6BhHu4'
```

```python
youtube = build(
    API_SERVICE_NAME,
    API_VERSION,
    developerKey=API_KEY,
)

request = youtube.commentThreads().list(
    part='snippet',
    videoId=VIDEO_ID,
    maxResults=100, # range 1-100
)

response = request.execute()

comments = []
for item in response['items']:
  data = item['snippet']['topLevelComment']['snippet']
  comments.append([
      data['authorDisplayName'],
      data['publishedAt'],
      data['textOriginal'],
  ])

while True:
  nextPageToken = response['nextPageToken']

  nextRequest = youtube.commentThreads().list(
      part="snippet",
      videoId=VIDEO_ID,
      maxResults=100, # range 1-100
      pageToken=nextPageToken,
  )

  response = nextRequest.execute()

  for item in response['items']:
    data = item['snippet']['topLevelComment']['snippet']
    comments.append([
        data['authorDisplayName'],
        data['publishedAt'],
        data['textOriginal'],
    ])
```

```python
import pandas as pd

pd.set_option('display.max_colwidth', None)

df = pd.DataFrame(
    comments,
    columns=['author', 'created_at', 'comment'],
)

df.shape
```
